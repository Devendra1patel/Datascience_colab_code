1. Line Plot

Use When:

To show trends over time.

To visualize continuous data.
Insights:

Shows upward trend in revenue.

Helps in forecasting future growth.

````python
import matplotlib.pyplot as plt

x = [2018, 2019, 2020, 2021, 2022]
y = [50, 60, 80, 100, 120]

plt.plot(x, y, marker='o', linestyle='-', color='blue', label='Sales Growth')
plt.xlabel("Year")
plt.ylabel("Revenue ($K)")
plt.title("Annual Sales Growth")
plt.legend()
plt.grid(True)
plt.show()
````

2. Bar Chart

Use When:

Comparing different categories.

Showing ranking or differences.
Insights:

Product B had the highest sales.

Product C underperformed compared to others.

````python
categories = ['Product A', 'Product B', 'Product C']
values = [100, 250, 150]

plt.bar(categories, values, color=['red', 'green', 'blue'])
plt.xlabel("Products")
plt.ylabel("Sales ($K)")
plt.title("Product Sales Comparison")
plt.show()
````

3. Histogram

Use When:

Understanding distribution of continuous data (e.g., exam scores, age distribution).
Insights:

Identifies how data is distributed.

Shows skewness or normal distribution.

````python
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=30, color='purple', edgecolor='black')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Distribution of Data")
plt.show()
````

4. Scatter Plot
Use When:
Finding relationships between two numerical variables.
Insights:
A positive correlation between X and Y values.
Helps in regression analysis.

````python
x = [1, 2, 3, 4, 5]
y = [5, 10, 15, 20, 25]

plt.scatter(x, y, color='blue', marker='o')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot Example")
plt.show()
````

5. Pie Chart
Use When:
Showing proportions or percentage breakdown.
Insights:
Apple holds the largest share.
Banana and Orange contribute equally.

````python
labels = ['Apple', 'Banana', 'Grapes', 'Orange']
sizes = [30, 25, 20, 25]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Fruit Distribution")
plt.show()

````

6. Box Plot

Use When:

Checking for outliers and data distribution.

Insights:

Identifies median and outliers.

Shows data spread.

````python
data = [7, 8, 9, 10, 15, 30, 50, 100, 120]

plt.boxplot(data)
plt.title("Box Plot Example")
plt.show()
````

7. Heatmap (Using Seaborn)

Use When:

Understanding correlation between multiple variables.
Insights:

High correlation areas are highlighted.

Helps in feature selection.

````python
import seaborn as sns
import numpy as np

data = np.random.rand(5,5)
sns.heatmap(data, annot=True, cmap="coolwarm")
plt.title("Heatmap Example")
plt.show()
````

8. Area Chart

Use When:

Similar to a line chart but used to show cumulative data.

Insights:

Shows total contribution over time.

````python
years = [2018, 2019, 2020, 2021, 2022]
values = [5, 15, 25, 40, 60]

plt.fill_between(years, values, color="skyblue", alpha=0.4)
plt.plot(years, values, color="blue", linestyle="dashed")
plt.xlabel("Year")
plt.ylabel("Growth")
plt.title("Area Chart Example")
plt.show()
````

````python
import numpy as np

categories = ['A', 'B', 'C', 'D']
values1 = [10, 15, 7, 20]
values2 = [5, 8, 12, 9]

plt.bar(categories, values1, color='blue', label="Category 1")
plt.bar(categories, values2, bottom=values1, color='orange', label="Category 2")

plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Stacked Bar Chart Example")
plt.legend()
plt.show()
````

````python
x = [10, 20, 30, 40, 50]
y = [15, 25, 35, 45, 55]
size = [100, 300, 500, 700, 1000]

plt.scatter(x, y, s=size, alpha=0.5, color="green")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bubble Chart Example")
plt.show()
````

````python
import seaborn as sns
import numpy as np

data = np.random.randn(100, 3)

sns.violinplot(data=data)
plt.title("Violin Plot Example")
plt.show()
````

````python
theta = np.linspace(0, 2*np.pi, 10)
r = np.random.rand(10)

plt.polar(theta, r, marker='o')
plt.title("Polar Chart Example")
plt.show()
````

````python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.random.rand(10)
y = np.random.rand(10)
z = np.random.rand(10)

ax.scatter(x, y, z, c='r', marker='o')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
plt.title("3D Scatter Plot Example")
plt.show()
````

````python
x = np.arange(10)
y = np.random.randint(1, 10, size=10)

plt.step(x, y, where='mid', label="Step Chart")
plt.xlabel("Time")
plt.ylabel("Value")
plt.title("Step Chart Example")
plt.legend()
plt.show()
````

````python
x = np.linspace(1, 10, 10)
y = np.sin(x)
error = np.random.rand(10) * 0.2

plt.errorbar(x, y, yerr=error, fmt='o', capsize=5, label="Error Bars")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Error Bar Example")
plt.legend()
plt.show()
````

### **Use When:**
- To show trends over time.
- To visualize continuous data.

```python

plt.grid(True)
plt.show()
```
